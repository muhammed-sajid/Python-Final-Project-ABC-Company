# Replace 'height' column with random values between 150 and 180

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('ABC company.csv.csv')
df['height'] = np.random.randint(150, 181, size=len(df))


mean = df['Salary'].mean()
median = df['Salary'].median()


df['Salary'] = df['Salary'].fillna(median)
df.isnull().sum()
```

#Task 1#
 Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
Team_distribution_percentage = df['Team'].value_counts(normalize=True) * 100
print('Team Distribution Percentage')
Team_distribution_percentage

```

Graphical Representation:

```python
#Plot for employee distribution
plt.figure(figsize=(12, 6))
Team_distribution_percentage.plot(kind='bar',color='Skyblue', alpha=0.7)

#Customize the plot
plt.title('Team Distribution by Percentage',fontsize=14)
plt.xlabel('Team',fontsize=12)
plt.ylabel('Percentage',fontsize=12)
plt.xticks(rotation=60, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

```python
Team_distribution_count = df['Team'].value_counts()
print('Team Distribution Count')
Team_distribution_count
```

Graphical Representation:

```python
#Prepare data for Seaborn
Team_distribution_count_df= Team_distribution_count.reset_index()
Team_distribution_count_df.columns = ['Team', 'Count']

#Plot for employee distribution
plt.figure(figsize=(12, 6))
sns.lineplot(data=Team_distribution_count_df, x='Team',y='Count', marker='o',color='blue', linewidth=2)

#Customize the plot
plt.title('Team Distribution by Count', fontsize=14)
plt.xlabel('Team',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=60, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

#Task 2#
Segregate employees based on their positions within the company.

```python
# Count the distribution of employees by position

position_distribution =df['Position'].value_counts()

position_distribution_df = position_distribution.reset_index()
position_distribution_df.columns =['Position', 'Count']

position_distribution_df
```

Graphical Representation:

```python
#plot using seaboen
plt.figure(figsize=(12, 6))
sns.barplot(data=position_distribution_df, x='Position', y='Count',hue='Position', palette='viridis')

# Plotting the distribution of positions
plt.title('Distribution of Employees by Position', fontsize=14)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=60, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

#Task 3#
 Identify the predominant age group among employees.

```python
# Step 1: Create age groups (bins)
bins = [10, 20, 30,40]
Label = ['10-20','20-30','30-40']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=Label, right=True)

# Step 2: Count the number of employees in each age group
Age_group_distribution = df['Age_Group'].value_counts()
Age_group_distribution
```

Graphical Representation:

```python
#plot using seaboen
plt.figure(figsize=(12, 6))
Age_group_distribution.plot.pie(
      autopct='%1.1f%%',
      startangle=90,
      colors=sns.color_palette('pastel'),
      labels=Age_group_distribution.index,
      wedgeprops={'edgecolor': 'black'}

)

plt.title('Predominant Age Group Among Employee', fontsize=16)
plt.ylabel('') # Remove the default ylabel
plt.tight_layout()

plt.show()
```

#Task 4#
Discover which team and position have the highest salary expenditure.

```python
# Group by Team and Position and calculate the total salary
Team_position_salary = df.groupby(['Team','Position'])['Salary'].sum().reset_index()

#Pivot the data for heatmap visualization
salary_pivot = Team_position_salary.pivot(index='Team', columns='Position', values='Salary')

#create heatmap to visualize salary expenditure
plt.figure(figsize=(12, 8))
sns.heatmap(salary_pivot, annot=True, fmt='.0f', cmap='Blues', linewidths=0.5, cbar_kws={'label': 'Total Salary'})

# Customize the plot
plt.title('Salary Expenditure by Team and Position', fontsize=16)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Team', fontsize=12)
plt.tight_layout()

plt.show()

#Identify the team and position with the highest salary expenditure
highest_salary = Team_position_salary.loc[Team_position_salary['Salary'].idxmax()]
print(f"The Team '{highest_salary['Team']}' ans Position '{highest_salary['Position']}' have the highest salary expenditure of {highest_salary['Salary']}  ")
```

#Task 5#
 Investigate if there's any correlation between age and salary, and represent it visually.

```python
#Calculate Correlation Coefficient

correlation = df['Age'].corr(df['Salary'])
print(f"Correlation between Age and Salary: {correlation:.2f}")

#Scatter Plot with Regression Line
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='Age', y='Salary', scatter_kws={'alpha':0.7}, line_kws={'color':'red'})

#Customize the plot
plt.title('Correlation Between Age and Salary', fontsize=14)
plt.xlabel('Age',fontsize=12)
plt.ylabel('Salary',fontsize=12)
plt.grid(axis='y', linestyle='--',alpha=0.7)
plt.tight_layout()

plt.show()
```
